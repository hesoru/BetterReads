# Data Cleaning and Wrangling for MongoDB database population

### data source: goodbooks-10k & goodreads_data_genre.csv

```python
import pandas as pd
```

```python
gb_books_df = pd.read_csv("Dataset_Options/goodbooks-10k/books.csv")
gb_books_df
```

```python
# list all the columns of the goodbook
print(gb_books_df.columns)
```

```python
gb_books_df.info()
```

```python
gb_books_cleaned_df = gb_books_df[["isbn", "authors", "original_publication_year", "title", "average_rating", 'ratings_count', 'work_text_reviews_count', 'books_count',"image_url"]]
gb_books_cleaned_df.head()
```

```python
gb_books_genre_description_df = pd.read_csv("Dataset_Options/goodreads_data_genre.csv")
gb_books_genre_description_df
```

```python
gb_books_genre_description_df.columns
```

```python
gb_books_genre_description_cleaned_df = gb_books_genre_description_df[['Book', 'Author', 'Description', 'Genres', 'Avg_Rating',
       'Num_Ratings']]
gb_books_genre_description_cleaned_df.head()
```

```python
# drop the duplicate rows
left = gb_books_cleaned_df.copy()
right = gb_books_genre_description_cleaned_df.copy()

# Lowercase columns
left['title_lower'] = left['title'].str.lower()
right['book_lower'] = right['Book'].str.lower()

# Drop duplicates
left = left.drop_duplicates(subset='title_lower')
right = right.drop_duplicates(subset='book_lower')

# Merge
merged_df = pd.merge(left, right, left_on='title_lower', right_on='book_lower', how='inner')

# Clean up
merged_df.drop(columns=['title_lower', 'book_lower'], inplace=True)
```

```python
merged_df
```

```python
merged_df.columns
```

```python
# Finalized the book data for better_reads
better_reads_df = merged_df[['title','authors','original_publication_year','image_url','Description','Genres','isbn', 
       'books_count','average_rating', 'ratings_count','work_text_reviews_count']]
better_reads_df
```

```python
# change the column names
better_reads_df = better_reads_df.rename(columns={
    'title': 'title',
    'authors': 'author',
    'original_publication_year': 'publishYear',
    'image_url': 'image',
    'Description': 'description',
    'Genres': 'genre',
    'isbn': 'ISBN',
    'books_count': 'numberOfEditions',
    'average_rating': 'averageRating',
    'ratings_count': 'ratingsCount',
    'work_text_reviews_count': 'reviewCount'
})
```

```python
better_reads_df
```

```python
better_reads_df.info()
```

```python
# drop rows with NA value
better_reads_df = better_reads_df.dropna()
```

```python
# drop rows with NA value
better_reads_df
```

```python
# convert year to int
better_reads_df['publishYear'] = better_reads_df['publishYear'].astype(int)
```

```python
better_reads_df
```

```python
better_reads_df.info()
```

```python
better_reads_df = better_reads_df.drop_duplicates(subset='ISBN', keep='first')
```

```python
# change the genre type to list of string
import ast

better_reads_df["genre"] = better_reads_df["genre"].apply(lambda x: ast.literal_eval(x))
```

```python
better_reads_df
```

```python
# exported to json file
better_reads_df.to_json("better_reads.json", orient="records", indent=4, force_ascii=False)
```
